```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path="/content/drive/MyDrive/Churn_Modelling.csv"
df=pd.read_csv(path)
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
print(df.dtypes)
```

```python
df['EstimatedSalary'] = df['EstimatedSalary'].astype(int)
```

```python
df['Age'] = df['Age'].astype(float)
```

```python
df['Exited'] = df['Balance'].astype(bool)
```

```python
print(df.dtypes)
```

```python
cat_cols = ['RowNumber', 'CustomerId', 'Age', 'Balance', 'Gender', 'NumOfProducts']

df_enc = pd.get_dummies(df, columns=cat_cols, drop_first=True)

```

```python
# Drop non-numeric columns explicitly
df_corr = df_enc.select_dtypes(include=[np.number])

# Now compute and plot the correlation matrix
plt.figure(figsize=(12,10))
corr = df_corr.corr()
sns.heatmap(corr, cmap='coolwarm', center=0, annot=False)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')

plt.subplot(1,2,2)
sns.histplot(df['Balance'], bins=30, kde=True)
plt.title('Balance')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Balance', y='Age', data=df)
plt.title('Age by Passenger Class')
plt.show()
```
